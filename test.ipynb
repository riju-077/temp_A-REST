Pro-Tips for the Amazon API
Pagination: If the response contains a nextToken, you must make the call again and include that nextToken in your next request body to get the second page of results.

Environment: If you are testing in the Sandbox environment, change the URL from advertising-api.amazon.com to advertising-api-test.amazon.com.

JSON Serialization: Note that campaignIdFilter and nameFilter expect objects with an "include": [] list inside them, not just a flat list of IDs.

Handling the Token (The "Refresh" Flow)
Since Amazon Access Tokens expire every 3,600 seconds (1 hour), you usually need to swap a refresh_token for a new access_token automatically.

```python
import requests
import json

def list_sponsored_products_campaigns(access_token, client_id, profile_id, filters=None):
    """
    Lists Amazon Sponsored Products campaigns using the v3 API.
    """
    url = "https://advertising-api.amazon.com/sp/campaigns/list"
    
    # Custom headers required by Amazon
    headers = {
        "Authorization": f"Bearer {access_token}",
        "Amazon-Advertising-API-ClientId": client_id,
        "Amazon-Advertising-API-Scope": profile_id,
        "Content-Type": "application/vnd.spCampaign.v3+json",
        "Accept": "application/vnd.spCampaign.v3+json"
    }

    # Default payload structure
    payload = {
        "maxResults": 100,
        "stateFilter": {
            "include": ["ENABLED", "PAUSED"]
        },
        "includeExtendedDataFields": True
    }

    # Merge custom filters if provided
    if filters:
        payload.update(filters)

    try:
        response = requests.post(url, headers=headers, json=payload)
        
        # Raise an exception for 4xx or 5xx errors
        response.raise_for_status()
        
        return response.json()

    except requests.exceptions.HTTPError as err:
        print(f"HTTP Error: {err.response.status_code} - {err.response.text}")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Example Usage ---
# auth_data = {
#     "access_token": "Atza|...",
#     "client_id": "amzn1.application-oa2-client...",
#     "profile_id": "1234567890"
# }
# 
# campaigns = list_sponsored_products_campaigns(**auth_data)
# print(json.dumps(campaigns, indent=2))
```

```python
import requests

AMAZON_URL = "https://advertising-api.amazon.com/sp/campaigns/list"

# ---- REQUIRED HEADERS ----
headers = {
    "Authorization": "Bearer YOUR_ACCESS_TOKEN",
    "Amazon-Advertising-API-ClientId": "YOUR_CLIENT_ID",
    "Amazon-Advertising-API-Scope": "YOUR_PROFILE_ID",
    "Content-Type": "application/vnd.spCampaign.v3+json"
}

# ---- REQUEST BODY ----
payload = {
    "includeExtendedDataFields": True,
    "maxResults": 10,
    "stateFilter": {
        "include": ["ENABLED"]
    }
}

# ---- API CALL ----
response = requests.post(
    AMAZON_URL,
    headers=headers,
    json=payload
)

# ---- RESPONSE ----
print("Status Code:", response.status_code)
print("Response JSON:")
print(response.json())
```
